## JOB CLASSIFICATION

```python
# variables
top_categories = ['Business Development (Sales)', 'Content Writing', 'Graphic Design', 'Marketing', 'Web Development', 'Digital Marketing']
classifier = pipeline("zero-shot-classification")

# function
def run_classifier(text):
  intermediate_results = classifier(text, candidate_labels=top_categories)
  return {i: v for i, v in zip(intermediate_results['labels'],intermediate_results['scores'])}
```

```python
# gradio objects
textbox = gr.Textbox(label="Write Job Responsbilities:", placeholder="Selected intern's day-to-day responsibilities include:", lines=2)

# gradio interface
interface = gr.Interface(
    run_classifier,
    inputs=textbox,
    outputs="label",
    theme="huggingface",
    title="Zero Shot Job Category Recognition",
    description="This tool can classify a whole internship description or responsbilities \
    statement from int these 6 job categories: \n"+",".join(top_categories),
    article="<p style='text-align: center'> Zero Shot Recognition | Demo Model</p>"
)
interface.launch(share=True)
```

## JOB SENTENCE COMPLETION

```python
# model
generator = pipeline("text-generation", max_length=30, stop_sequence="\n")

# function
def run_generator(text):
  intermediate_results = generator(text)
  return intermediate_results[0]['generated_text']
```

```python
# gradio objects
textbox = gr.Textbox(label="Write Job Responsbilities:", placeholder="Selected intern's day-to-day responsibilities include:", lines=2)
textbox_output = gr.Textbox(label="Generated Sequence", placeholder="Wait for the magic.", lines=2)

# gradio interface
interface = gr.Interface(
    run_generator,
    inputs=textbox,
    outputs=textbox_output,
    theme="huggingface",
    title="JOB SENTENCE COMPLETION",
    description="This tool can complete a given line of job responsibilities.\n",
    article="<p style='text-align: center'> Zero Shot Recognition | Pipeline Text Generation </p>"
)
interface.launch(share=True)
```

##JOB SUMMARIZATION

```python
# model
summarizer = pipeline("summarization", model="facebook/bart-large-cnn", max_length=150)
```

```python
# function
def run_summarizer(text):
    summary = summarizer(text)
    return summary[0]['summary_text']
```

```python
# gradio objects
textbox_summarizer = gr.Textbox(label="Input Text:", placeholder="Paste your text here.", lines=10)
textbox_output_summarizer = gr.Textbox(label="Generated Summary", placeholder="Wait for the summary.", lines=5)
```

```python
# gradio interface
interface_summarizer = gr.Interface(
    run_summarizer,
    inputs=textbox_summarizer,
    outputs=textbox_output_summarizer,
    theme="huggingface",
    title="JOB SUMMARIZATION",
    description="Generate summaries for given job description using the BART model.",
    article="<p style='text-align: center'> Summarization | Pipeline Text Generation </p>"
)
interface_summarizer.launch(share=True)
```
